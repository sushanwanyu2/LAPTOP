```python
# 冒泡排序
```

```python
def sort(numbers):
    for i in range(0,len(numbers)-1):
        for j in range(0,len(numbers)-1-i):
            if numbers[j]<numbers[j+1]:
                numbers[j],numbers[j+1]=numbers[j+1],numbers[j]
    return numbers
```

```python
list1=[1,34,563,4,36,356,67,4784,45]
```

```python
sort(list1)
```
